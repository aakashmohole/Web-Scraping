```python
from autoscraper import AutoScraper
url = "https://www.flipkart.com/search?q=samsung"
wanted = ["₹14,499","SAMSUNG Galaxy F15 5G (Groovy Violet, 128 GB)","14% off"]
```

```python
scraper = AutoScraper()
result = scraper.build(url,wanted)
```

```python
result
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_2t26':'Price','rule_91ui':'Title','rule_o7yw':'Discount'})
scraper.keep_rules(['rule_2t26','rule_91ui','rule_o7yw'])
scraper.save('flipkart_search')
```

```python
result = scraper.get_result_similar("https://www.flipkart.com/search?q=samsung",group_by_alias=True)
```
